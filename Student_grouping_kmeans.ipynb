```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
cgpa = np.random.uniform(1.0, 4.0, 100)
iq = np.random.uniform(90, 130, 100)
```

```python
df = pd.DataFrame({'CGPA': cgpa, 'IQ': iq})
```

```python
km = KMeans(n_clusters=2, n_init=10)
df['Group_Sticker'] = km.fit_predict(df)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['CGPA'], df['IQ'], c=df['Group_Sticker'], cmap='viridis')
```

```python
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
            color='red', marker='X', s=200, label='Group Centers')

plt.title("Student Groups Identified by AI (Unsupervised)")
plt.xlabel("CGPA (4.0 Scale)")
plt.ylabel("IQ")
plt.legend()
plt.show()
```
